Importing all the libraries

```
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

Reading the data

```
data = pd.read_csv('http://bit.ly/w-data')
```

Printing the first few files

```
data.head(5)
```

Checking for a relationship between Hours and Scores variable

```
plt.figure(figsize=(16,8))
plt.scatter(data['Hours'],data['Scores'],c='red')
plt.xlabel("Hours")
plt.ylabel("Scores")
plt.show()
```

Dividing the data into attributes and labels

```
x = data['Hours'].values.reshape(-1,1) #attribute
y = data['Scores'].values.reshape(-1,1) #label which is to be predicted
```

```
from sklearn.model_selection import train_test_split

# split data into 80% for training and 20% for test set
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```
regressor = LinearRegression()  
regressor.fit(x_train, y_train)
print("Intercept is :" , regressor.intercept_)
print("Slope is:",regressor.coef_)
```

```
y_pred = regressor.predict(x_test)
y_pred
```

```
df = pd.DataFrame({'Actual': y_test.flatten(), 'Predicted': y_pred.flatten()})
df
```

```
# plotting the graph to visualize the actual and predicted in a better way
#selecting first 10 records
df_5 = df.head(5)
df_5.plot(kind='bar',figsize=(15,7))
plt.grid(which='major',linestyle='-',linewidth='0.5',color='green')
plt.grid(which='minor',linestyle=':',linewidth='0.5',color='black')
plt.show()
```

```
# prediction if the student studies for 9.25 hours
regressor.predict([[9.25]])
```
